```python
#Standard
import numpy as np
import pandas as pd
from numpy.random import randn

# Stats imports
from scipy import stats

#Plotting imports
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1,data2],vert=True)
```

```python
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.boxplot(data1,vert=False)
```

```python
#Normal Dist
data1 = stats.norm(0,5).rvs(100)

#Two gamma distributions concatenated
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                      -1*stats.gamma(5).rvs(50)])

sns.boxplot([data1,data2],whis=np.inf)

#Violin plot - combine a kde plot with a box plot
```

```python
sns.violinplot([data1,data2],vert=True)
```

```python
sns.violinplot(data2,bw=0.01,vert=True) #small bandwidth shows sharp variations
```

```python
sns.violinplot(data1,inner='stick',vert=True)
```
