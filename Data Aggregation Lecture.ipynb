```python
"""The data we will use for this course is the archives of the UC Erwine Machine Learning class"""
"""check out this link, we will use the winequality-red.csv from that site for our analysis"""
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality'
```

```python
"""Get the file form that site and download to notebooks folks
the separator on that file is ; but when saved as csv, may change to comma, so please
use comma as sep below as below"""
dframe_wine = pd.read_csv('winequality-red.csv',sep=',')
```

```python
dframe_wine['alcohol'].mean()
```

```python
dframe_wine
```

```python
def max_to_min(arr): 
    return arr.max() - arr.min()
```

```python
wino = dframe_wine.groupby('quality')
```

```python
wino.describe()
```

```python
"""the max_to_min function we wrote above is an aggregation function. 
This function is passed as a parameter to the aggregation function on the dataframe so that the aggregation 
operation can be carried out"""
wino.agg(max_to_min)
```

```python
wino.agg('mean')
```
