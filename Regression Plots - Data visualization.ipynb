```python
#Standard
import numpy as np
import pandas as pd
from numpy.random import randn

# Stats imports
from scipy import stats

#Plotting imports
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
"""Seaborn comes with some example datasets which will use"""
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head
```

```python
#lmplot gives a regression plot
#here we have plotted total_bill column vs tip column on the tips dataframe
#it produced a scatter plot of values and a linear regression plot
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,order=4, #4th order fit
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg='False') #get rid of regression plot
```

```python
tips.head()
```

```python
tips['tip_pect'] = 100*(tips['tip']/tips['total_bill']) #new column tips_pect
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
#now use Jitter. read about Jitter on wikipedia
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter=1)
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
"""Local regression or Lowess"""
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot('total_bill','tip_pect',tips) #regplot is the lower level function that lmplot is built on top of
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)
sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(tips['tip_pect'],tips['size'],color='red',ax=axis2,bw=10)
```
